```python
n = gp.read_file('/home/nicolas/LSS/IFC/Nicolas/iowa_hd/with_nhd_high/merged_BurnLines_edited.shp')
```

```python
n['log_area'] = n['DivDASqKm'].apply(lambda x: np.log(x))
n['depth'] = (n['log_area']+10)*2
n.loc[np.isinf(n['depth']), 'depth'] = 10
n.loc[n['depth'] < 10, 'depth'] = 10
```

```python
n.to_file('/home/nicolas/LSS/IFC/Nicolas/iowa_hd/with_nhd_high/merged_BurnLines_edited.shp')
```
